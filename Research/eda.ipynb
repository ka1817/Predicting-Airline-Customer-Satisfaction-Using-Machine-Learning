```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='satisfaction', hue='Gender', palette='coolwarm')

plt.title("Satisfaction Levels by Gender")
plt.xlabel("Satisfaction")
plt.ylabel("Count")
plt.legend(title="Gender")
plt.show()
```

```python
gender_satisfaction = df.groupby(['Gender', 'satisfaction']).size().unstack()
gender_satisfaction = gender_satisfaction.div(gender_satisfaction.sum(axis=1), axis=0) * 100  # Convert to %

gender_satisfaction.plot(kind='bar', stacked=True, figsize=(8, 5), colormap='coolwarm')

plt.title("Satisfaction Percentage by Gender")
plt.xlabel("Gender")
plt.ylabel("Percentage")
plt.legend(title="Satisfaction")
plt.show()
```

```python
for gender in df['Gender'].unique():
    subset = df[df['Gender'] == gender]['satisfaction'].value_counts()
    
    plt.figure(figsize=(5, 5))
    plt.pie(subset.values, labels=subset.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
    plt.title(f"Satisfaction Distribution for {gender}")
    plt.show()
```

```python
# Create a contingency table
import scipy.stats as stats

contingency_table = pd.crosstab(df['Gender'], df['satisfaction'])

# Perform Chi-Square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Display results
print("Chi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-Value: {p:.4f}")

# Interpret results
alpha = 0.05  # 95% confidence level
if p < alpha:
    print("Conclusion: There is a significant relationship between Gender and Satisfaction (Reject Null Hypothesis)")
else:
    print("Conclusion: No significant relationship between Gender and Satisfaction (Fail to Reject Null Hypothesis)")
```

Key Takeaways

✔ Visual insight: Males and females show similar satisfaction trends, but females appear slightly more dissatisfied.

✔ Statistical confirmation: The difference is significant, meaning gender does influence satisfaction levels.

```python
import scipy.stats as stats
import pandas as pd

alpha = 0.05 

for col in categorical_features:
    if col != 'satisfaction':  
        print(f"\nChi-Square Test for {col} vs Satisfaction")
        contingency_table = pd.crosstab(df[col], df['satisfaction'])
        chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
        
        print(f"Chi-Square Statistic: {chi2:.4f}")
        print(f"Degrees of Freedom: {dof}")
        print(f"P-Value: {p:.4f}")
        
        if p < alpha:
            print("Conclusion: Significant relationship (Reject Null Hypothesis)")
        else:
            print("Conclusion: No significant relationship (Fail to Reject Null Hypothesis)")
```

## Summary of Conclusions from the Analysis

Based on the Chi-Square tests and countplots, we can draw the following conclusions:

✅Gender and Satisfaction

Statistical Finding: The Chi-Square test result (p-value = 0.0001) indicates a significant relationship between gender and satisfaction.
Visual Insight: From the countplot, males and females show similar satisfaction distributions, though males have slightly higher neutral/dissatisfied counts.
Conclusion: Gender does influence satisfaction, but the effect size appears to be small compared to other factors.

✅Customer Type and Satisfaction

Statistical Finding: The p-value is 0.0000, meaning a highly significant relationship exists.
Visual Insight: Loyal customers have a much higher satisfaction rate than disloyal customers.
Conclusion: Customer loyalty is strongly correlated with satisfaction. Loyal customers are much more likely to be satisfied.


✅Type of Travel and Satisfaction

Statistical Finding: The p-value is 0.0000, indicating a strong relationship between travel type and satisfaction.
Visual Insight: Business travelers show higher satisfaction, while personal travelers have a much lower satisfaction rate.
Conclusion: Business travelers are significantly more satisfied compared to personal travelers. Airlines may prioritize business travelers, leading to better service quality.

✅Class and Satisfaction

Statistical Finding: The p-value is 0.0000, showing a highly significant relationship between travel class and satisfaction.
Visual Insight: Business class passengers have the highest satisfaction, while Economy class has the lowest satisfaction rate.
Conclusion: Travel class strongly affects satisfaction. Passengers in Business Class are much more likely to be satisfied, whereas Economy Class travelers report lower satisfaction levels.



Final Takeaways

✅ Customer Type, Type of Travel, and Class have the strongest impact on satisfaction.

✅ Gender has a statistically significant effect, but it is less impactful than other factors.

✅ Airlines should focus on improving service for personal travelers and Economy Class passengers to boost overall satisfaction.

```python
# Percentage distribution of categorical features
categorical_features = ['Gender', 'Customer Type', 'Type of Travel', 'Class']
for feature in categorical_features:
    print(f"\n{feature} Distribution:")
    print(df[feature].value_counts(normalize=True) * 100)
```

```python
from scipy.stats import chi2_contingency

# Perform chi-square tests between categorical variables
category_pairs = [('Gender', 'Customer Type'), ('Class', 'Type of Travel'), ('Type of Travel', 'Customer Type')]

for cat1, cat2 in category_pairs:
    contingency_table = pd.crosstab(df[cat1], df[cat2])
    chi2, p, dof, _ = chi2_contingency(contingency_table)
    print(f"\nChi-Square Test: {cat1} vs {cat2}")
    print(f"Chi-Square Statistic: {chi2:.4f}, P-Value: {p:.4f}")
    print("Conclusion:", "Significant Relationship" if p < 0.05 else "No Significant Relationship")
```

Overall Key Takeaways

1️⃣ Gender impacts customer loyalty. Airlines should tailor loyalty programs for different genders.

2️⃣ Class choice is related to travel purpose. Airlines can optimize class pricing and marketing.

3️⃣ Business travelers are more loyal. Special incentives should be created to encourage repeat personal travel.

```python
##Feature Importence among Catgegorical features

from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()
for col in categorical_features:
    df_encoded[col] = LabelEncoder().fit_transform(df[col])

X = df_encoded[categorical_features]
y = df_encoded['satisfaction']

model = DecisionTreeClassifier()
model.fit(X, y)

# Plot Feature Importance
importances = pd.Series(model.feature_importances_, index=categorical_features)
importances.sort_values().plot(kind="barh", colormap="coolwarm")
plt.title("Feature Importance (Decision Tree)")
plt.show()
```

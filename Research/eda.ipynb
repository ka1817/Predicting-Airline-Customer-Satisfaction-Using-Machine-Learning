```python
from scipy.stats import ttest_ind, mannwhitneyu
import seaborn as sns
import matplotlib.pyplot as plt

# Check Normality using Shapiro-Wilk test
from scipy.stats import shapiro

satisfied_scores = df[df['satisfaction'] == 'satisfied']['Total_Service_Score']
dissatisfied_scores = df[df['satisfaction'] == 'neutral or dissatisfied']['Total_Service_Score']

# Normality test
shapiro_satisfied = shapiro(satisfied_scores.sample(5000, random_state=42)) if len(satisfied_scores) > 5000 else shapiro(satisfied_scores)
shapiro_dissatisfied = shapiro(dissatisfied_scores.sample(5000, random_state=42)) if len(dissatisfied_scores) > 5000 else shapiro(dissatisfied_scores)

print(f"Shapiro-Wilk Test (Satisfied): p-value = {shapiro_satisfied.pvalue}")
print(f"Shapiro-Wilk Test (Dissatisfied): p-value = {shapiro_dissatisfied.pvalue}")

# Choose test based on normality
if shapiro_satisfied.pvalue > 0.05 and shapiro_dissatisfied.pvalue > 0.05:
    # If normally distributed, use t-test
    t_stat, p_value = ttest_ind(satisfied_scores, dissatisfied_scores, equal_var=False)
    print(f"T-test: p-value = {p_value}")
else:
    # If not normally distributed, use Mann-Whitney U test
    u_stat, p_value = mannwhitneyu(satisfied_scores, dissatisfied_scores, alternative='two-sided')
    print(f"Mann-Whitney U Test: p-value = {p_value}")

# Plot distributions
plt.figure(figsize=(10, 5))
sns.kdeplot(satisfied_scores, label="Satisfied", shade=True)
sns.kdeplot(dissatisfied_scores, label="Neutral/Dissatisfied", shade=True)
plt.legend()
plt.title("Distribution of Total_Service_Score by Satisfaction")
plt.show()
```

Yes, you are absolutely right! The distributions look visually normal, but the Shapiro-Wilk test p-values are extremely small, indicating that both groups are statistically non-normal. This is why we rely on statistical tests rather than just visual inspection.



The key takeaway: Always rely on statistical tests, not just skewness, to determine normality. 🚀



Conclusions from the Analysis

1️⃣ Customer Satisfaction is Strongly Linked to Service Quality

The Total_Service_Score is significantly higher for satisfied customers compared to neutral/dissatisfied ones.

The Mann-Whitney U test p-value is 0.0, confirming that this difference is statistically significant.

2️⃣ Dissatisfied Customers Experience Lower Service Scores

The distribution of dissatisfied customers is centered around lower service scores (30–50 range), while satisfied customers tend to have higher scores (above 50).

This suggests that lower service ratings are a strong predictor of dissatisfaction.

3️⃣ Service Quality is Not Normally Distributed

4️⃣ Higher Service Ratings Lead to a Greater Likelihood of Satisfaction

Dissatisfied customers show a more balanced distribution but with a peak at lower scores.

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer


ordinal_categories = [
    ['No Delay', 'Short Delay', 'Moderate Delay', 'Long Delay'], 
    ['Poor', 'Average', 'Good', 'Excellent']  
]

trf = ColumnTransformer([
    ('ohe', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), [0, 1, 3, 4]), 
    ('ordinal', OrdinalEncoder(categories=ordinal_categories), [6, 7]) 
], remainder='passthrough') 
```

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
trf2=GradientBoostingClassifier()
```

```python
from sklearn.pipeline import Pipeline
pipe=Pipeline([
    ('trf',trf),
    ('trf2',trf2),
])
```

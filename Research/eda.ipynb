```python
gender_satisfaction = df.groupby(['Gender', 'satisfaction']).size().unstack()
gender_satisfaction = gender_satisfaction.div(gender_satisfaction.sum(axis=1), axis=0) * 100  # Convert to %

gender_satisfaction.plot(kind='bar', stacked=True, figsize=(8, 5), colormap='coolwarm')

plt.title("Satisfaction Percentage by Gender")
plt.xlabel("Gender")
plt.ylabel("Percentage")
plt.legend(title="Satisfaction")
plt.show()
```

```python
for gender in df['Gender'].unique():
    subset = df[df['Gender'] == gender]['satisfaction'].value_counts()
    
    plt.figure(figsize=(5, 5))
    plt.pie(subset.values, labels=subset.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
    plt.title(f"Satisfaction Distribution for {gender}")
    plt.show()
```

```python
# Create a contingency table
import scipy.stats as stats

contingency_table = pd.crosstab(df['Gender'], df['satisfaction'])

# Perform Chi-Square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Display results
print("Chi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-Value: {p:.4f}")

# Interpret results
alpha = 0.05  # 95% confidence level
if p < alpha:
    print("Conclusion: There is a significant relationship between Gender and Satisfaction (Reject Null Hypothesis)")
else:
    print("Conclusion: No significant relationship between Gender and Satisfaction (Fail to Reject Null Hypothesis)")
```

Key Takeaways

✔ Visual insight: Males and females show similar satisfaction trends, but females appear slightly more dissatisfied.

✔ Statistical confirmation: The difference is significant, meaning gender does influence satisfaction levels.

Final Takeaways

✅ Customer Type, Type of Travel, and Class have the strongest impact on satisfaction.

```python
from scipy.stats import chi2_contingency

# Perform chi-square tests between categorical variables
category_pairs = [('Gender', 'Customer Type'), ('Class', 'Type of Travel'), ('Type of Travel', 'Customer Type')]

for cat1, cat2 in category_pairs:
    contingency_table = pd.crosstab(df[cat1], df[cat2])
    chi2, p, dof, _ = chi2_contingency(contingency_table)
    print(f"\nChi-Square Test: {cat1} vs {cat2}")
    print(f"Chi-Square Statistic: {chi2:.4f}, P-Value: {p:.4f}")
    print("Conclusion:", "Significant Relationship" if p < 0.05 else "No Significant Relationship")
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
X=df.drop('satisfaction',axis=1)
y=df['satisfaction']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer

delay_order = [["No Delay", "Minor Delay", "Moderate Delay", "Severe Delay"]]

trf = ColumnTransformer([
    ('ordinal_delay', OrdinalEncoder(categories=delay_order), [7]), 
    ('ohe_categorical', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), [0, 1, 3, 4])  
], remainder='passthrough')

```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
```

```python
pipe=Pipeline([
    ('trf',trf),
    ('classifier',classifier)
])
```

```python
pipe
```

```python
# Define Hyperparameter Grid
param_grid = {
    'classifier__n_estimators': [50, 100, 200],
    'classifier__max_depth': [10, 20, None],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4]
}
```

```python
grid_search = GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
```

```python
best_model=grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
from sklearn.metrics import classification_report
y_pred=best_model.predict(X_test)
print(classification_report(y_test,y_pred))
```
